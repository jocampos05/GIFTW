```python
data_europe = pandas.read_csv('gapminder_gdp_europe.csv', index_col='country')
years = data_europe.columns.str.strip('gdpPercap_') #cambio las columnas como a√±os. hago esto primero
years_int = years.astype(int) 
data_europe.columns = years_int
data_europe
```

```python
data_europe.min().plot(label='min',color = 'green', style = "--")
data_europe.min()
data_europe.max().plot(label='max', color = 'fuchsia')
data_europe.max()
mpl.xlabel('years')
mpl.ylabel('GDP per capita')
mpl.legend(loc='best')
```
